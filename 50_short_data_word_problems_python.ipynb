## Problem 15

**Question:** A website recorded visit durations (seconds): `dur = [35, 120, 58, 240, 15]`. Compute the median duration (assume odd length OR even length handled correctly).

*Hint:* Sort first, then handle odd/even length.

````python
# Your turn: write your solution here

````

<details>
<summary><strong>Show answer</strong></summary>

```python
dur = [35, 120, 58, 240, 15]
d = sorted(dur)
n = len(d)
mid = n // 2
if n % 2 == 1:
    med = d[mid]
else:
    med = (d[mid-1] + d[mid]) / 2
```
